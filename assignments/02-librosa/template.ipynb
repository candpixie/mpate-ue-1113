```python
import librosa
```

```python
# Load the audio file (replace with the path of your own file)
audio_file = 'Youth.aiff'
y, sr = librosa.load(audio_file)
```

```python
# Compute the tempo and beat frames
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
print('Estimated tempo: {:.2f} beats per minute'.format(tempo[0]))
```

```python
tensec = y[sr*60:sr*70]
```

```python
import IPython.display
IPython.display.Audio(tensec, rate=sr)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Draw the waveform of the 10-second segment
plt.figure(figsize=(14, 4))
librosa.display.waveshow(tensec, sr=sr)
plt.title('Waveform of the 10-second segment')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
onset_frames = librosa.onset.onset_detect(y=tensec, sr=sr, backtrack=True, delta=0.04, wait=4)
onset_times = librosa.frames_to_time(onset_frames, sr=sr)
```

```python
import librosa.display
plt.figure(figsize=(14, 4))
librosa.display.waveshow(tensec, sr=sr)
plt.title('Waveform of the 10-second segment with onsets')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.vlines(onset_times, -1, 1, color='k')
plt.show()
```
