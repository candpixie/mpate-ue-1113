```python
import numpy as np
from IPython.display import Audio
import matplotlib.pyplot as plt
```

```python
# CREATING A SINE WAVE

# THESE VALUES WILL BE USED TO COMPUTE ALL WAVES
duration = 5
sampleRate = 44100
frequency = 440
pi = 3.141592654

time = np.linspace(0,duration,sampleRate * duration)

sineWave = np.sin(2 * pi * time * frequency)
Audio(sineWave,rate=sampleRate)
```

```python
plt.plot(sineWave[0:500])
```

```python
# CONVERTING SINE WAVE TO SQUARE WAVE

square = sineWave

for i in range(len(sineWave)):
  if sineWave[i] > 0:
    square[i] = 1
  else:
    square[i] = -1

Audio(square, rate=sampleRate)
```

```python
plt.plot(square[0:500])
```

```python
# CREATING A SAWTOOTH WAVE

saw = sineWave
samplesPerCycle = int(sampleRate / frequency)

s = np.linspace(-1, 1, samplesPerCycle)
for i in range(0,len(sineWave),samplesPerCycle):
  for j in range(samplesPerCycle):
    saw[i+j] = s[j]


Audio(saw,rate=sampleRate)
```

```python
plt.plot(saw[0:500])
```
