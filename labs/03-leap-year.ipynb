```python
def leapYear(year):
    # Check if the year is evenly divisible by 4
    if year % 4 == 0:
        # Check if the year is also divisible by 100
        if year % 100 == 0:
            # Check if the year is also divisible by 400
            if year % 400 == 0:
                # If divisible by 400, it's a leap year
                return 1
            else:
                # If not divisible by 400, it's not a leap year
                return 0
        else:
            # If not divisible by 100, it's a leap year
            return 1
    else:
        # If not divisible by 4, it's not a leap year
        return 0

# Example usage
year = 2023
is_leap = leapYear(year)
print(f"Year {year} is a leap year: {bool(is_leap)}")
```
