## 16) A Tiny Music Example: Convert MIDI Numbers to Note Names
This is a small, classic music-programming task.


```python
NOTE_NAMES = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']

def midi_to_name(midi_number):
    octave = (midi_number // 12) - 1
    name = NOTE_NAMES[midi_number % 12]
    return f"{name}{octave}"

for m in [60, 61, 64, 69, 72]:
    print(m, '→', midi_to_name(m))
```

## 18) Mini-Project: Text Beat Pattern Generator
We’ll represent a beat grid as text.


```python
def text_beat(kick_steps, snare_steps, steps=16):
    line_k = ['-'] * steps
    line_s = ['-'] * steps
    for i in kick_steps:
        if 0 <= i < steps:
            line_k[i] = 'K'
    for i in snare_steps:
        if 0 <= i < steps:
            line_s[i] = 'S'
    return 'Kick : ' + ''.join(line_k) + '\n' + 'Snare: ' + ''.join(line_s)

print(text_beat(kick_steps=[0, 8], snare_steps=[4, 12]))
```

### Example A: Generating a Sine Wave (Pure Tone)

```python
import numpy as np
import matplotlib.pyplot as plt

sample_rate = 44100
duration = 0.01
frequency = 440

t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
wave = np.sin(2 * np.pi * frequency * t)

plt.plot(t[:200], wave[:200])
plt.title('440 Hz Sine Wave (Zoomed In)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.show()
```

### Example C: Data → Interpretation

```python
bpms = [60, 72, 90, 120, 144]

for bpm in bpms:
    if bpm < 80:
        feel = 'slow / relaxed'
    elif bpm < 120:
        feel = 'moderate'
    else:
        feel = 'fast / energetic'
    print(bpm, '→', feel)
```
