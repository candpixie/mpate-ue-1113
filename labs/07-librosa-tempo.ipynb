```python
import librosa

def compute_tempo(audio_file):
    # Load the audio file
    y, sr = librosa.load(audio_file)
    
    # Compute the tempo and beat frames
    tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
    
    print(f"Tempo: {tempo} BPM")
    
    return tempo

# Replace 'path/to/your/audiofile.wav' with the path to the audio file you want to analyze
audio_file = '/Users/rowe/Documents/Youth.aiff'
compute_tempo(audio_file)
```
